```python
# Check number of classes
n_classes = len(set(data['label']))
print(f'Number of classes: {n_classes}')
n_img = len(data['label'])
print(f'Number of images in total: {n_img}')
```

```python
# Get number of occurences per label and plot
occur = Counter(data['label'])
occur = dict(sorted(occur.items(), key=lambda item: item[1]))
names = list(occur.keys())
values = list(occur.values())

plt.figure(figsize=(12, 4))
plt.xticks(rotation=45)
plt.bar(range(len(occur)), values, tick_label=names)
plt.show()
```

```python
# check size of images
rows = []
cols = []
for im in data['data']:
    row, col = im.shape
    rows.append(row)
    cols.append(col)

print(f'Mean dimensions of images: ({np.mean(rows):.2f}, {np.mean(cols):.2f})')
print(f'Max height: {max(rows)}, Min height: {min(rows)}')
print(f'Max width: {max(cols)}, Min width: {min(cols)}')
```
